# Create week and year columns in DataFrame

```
import pandas as pd
data['Date'] = pd.to_datetime(data['Date'])
data['Week'] = data['Date'].dt.isocalendar().week
data['Year'] = data['Date'].dt.year
```

```
data
```

```
data['week'] = data['Date'].dt.isocalendar().week
data['year'] = data['Date'].dt.isocalendar().year

# Calculate the total Amount by week and year
weekly_amount = data.pivot_table(values='Amount', index=['year', 'week'], aggfunc='sum').reset_index()

# Create a 'Week' column to easily plot the data
weekly_amount['Week'] = pd.to_datetime(weekly_amount[['year', 'week']].astype(str).agg('-'.join, axis=1) + '-1', format='%Y-%W-%w')

# Plotting the data
plt.figure(figsize=(15, 10))
plt.plot(weekly_amount['Week'], weekly_amount['Amount'])
plt.title('Total Amount by Week')
plt.xlabel('Week')
plt.ylabel('Total Amount')
plt.grid()
plt.xticks(rotation=360)  # Rotate x-axis labels for better readability
plt.show()
```

# Total Amount by Month and Category

```
# prompt: generate code for Total Amount by Month and Category

import matplotlib.pyplot as plt
# Calculate the total Amount by month and category
monthly_category_amount = data.groupby([data['Date'].dt.to_period('M'), 'Category'])['Amount'].sum().reset_index()

# Rename the 'Date' column to 'Month'
monthly_category_amount.rename(columns={'Date': 'Month'}, inplace=True)

# Display the result
print(monthly_category_amount)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Amount', hue='Category', data=monthly_category_amount)
plt.title('Total Amount by Month and Category')
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

* Blazers had the highest sales in March and May 2021.  
* Perfume sales were relatively consistent across all months.   
* Shoes and Socks had lower sales compared to other categories.
* Watches had moderate sales, peaking slightly in June 2021.

```
data['Date'] = pd.to_datetime(data['Date'])  # Convert to datetime if not already
heatmap_data = data.pivot_table(values='Amount', index='Category', columns=data['Date'].dt.to_period('M'), aggfunc='sum')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu")  # Use a suitable colormap
plt.title('Total Amount by Category and Month')
plt.xlabel('Month')
plt.ylabel('Category')
plt.show()
```

```
category_amounts = data.groupby('Category')['Amount'].sum()

# Create the pie chart
plt.figure(figsize=(8, 8))  # Adjust figure size as needed
plt.pie(category_amounts, labels=category_amounts.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
