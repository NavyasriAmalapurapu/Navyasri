```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

```
plt.figure(figsize=(12, 6))
df.groupby('date')['total_cases'].sum().plot()
plt.title('Global Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.show()
```

```
# Plot total deaths over time
plt.figure(figsize=(12, 6))
df.groupby('date')['total_deaths'].sum().plot(color='red')
plt.title('Global Total COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.show()
```

```
# Global Trends - COVID-19 cases and deaths
global_data = df.groupby('date').sum()

plt.figure(figsize=(14, 8))
plt.plot(global_data.index, global_data['total_cases'], label='Total Cases')
plt.plot(global_data.index, global_data['total_deaths'], label='Total Deaths', color='red')
plt.title('Global Trends: COVID-19 Cases and Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])


# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 7))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
# Compute the correlation matrix
correlation_matrix = numeric_df.corr()
```
